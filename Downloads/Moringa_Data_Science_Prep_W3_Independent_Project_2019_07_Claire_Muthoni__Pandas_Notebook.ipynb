Now we begin to clean the data for the cells database

```
# Trying to make the data more legible 
missing_values = [';','"','...']
something = '...'
Cells =pd.read_csv('cells_geo.csv',na_values=missing_values)
Cells
```

```
# We will now try to rename the column 
Cells_new = Cells.rename(columns={';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE': 'VILLES'},
                         inplace=False)
Cells_new
```

```
#We will now drop the first column
Cells_new.drop(columns=['VILLES'],inplace=True)
```

```
# Since the there is only one column and all the values of the second column are the same 
# Then we can make a dictionary for the columns and concatanate them  
addition = {'VILLES':['ADJAME','ADJAME','ADJAME','ADJAME','ASSINIE','ASSINIE','ASSINIE','ZUENOULA','ZUENOULA','ZUENOULA',],
            'Status': ['Inservice','Inservice','Inservice','Inservice','Inservice','Inservice','Inservice','Inservice','Inservice','Inservice'],
            'Localisation': ['ABIDJAN','ABIDJAN','ABIDJAN','INTERIEUR','INTERIEUR','INTERIEUR','INTERIEUR','INTERIEUR','INTERIEUR','INTERIEUR'],
            'DecoupZone':['Abidjan_EST','Abidjan_EST ','Abidjan_EST ','Grand-EST','Grand-EST','Grand-NORD','Grand-NORD','Grand-NORD','Grand-NORD','Grand-NORD']}
df = pd.DataFrame.from_dict(addition, orient='columns')
df
data = [Cells_new,df]
clean_Cells = pd.concat(data)
clean_Cells
```
